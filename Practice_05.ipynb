#### Approach 1: Calculating the min, max, and average in-state tution of the above four institutions. (Very inefficient way)

```
#Calculate the min, max, and average in-state tution of the above four institutions.
INSTNM_list = ['University of Maryland-Baltimore County','University of Maryland-College Park','Loyola University Maryland','Johns Hopkins University']
stats_list = []


for school in INSTNM_list:
  #print(school)
  temp_df = df_sm.loc[df_sm['INSTNM']== school ]
  temp_df = temp_df.reset_index()
  stats_list.append(temp_df.loc[0].at['TUITIONFEE_IN'])
  

sum_value =sum(stats_list)
min_value =min(stats_list)
max_value =max(stats_list)
avg_value =sum_value/len(stats_list)

print(stats_list)
#print(type(stats_list[1]))
print(f" The min value of in-state tution of the above four institutions: {min_value}")
print(f" The max value of in-state tution of the above four institutions: {max_value}")
print(f" The average value of in-state tution of the above four institutions: {avg_value}")
```

#### Approach 2: Calculating the min, max, and average in-state tution of the above four institutions. (Efficient way with pandas)

```
INSTNM_list = ['University of Maryland-Baltimore County','University of Maryland-College Park','Loyola University Maryland','Johns Hopkins University']

new_df = df_sm[df_sm.INSTNM.isin(INSTNM_list)]
tuition_min =new_df['TUITIONFEE_IN'].min()
tuition_max =new_df['TUITIONFEE_IN'].max()
tuition_avg = new_df['TUITIONFEE_IN'].mean()


print(f" The min value of in-state tution of the above four institutions: {tuition_min}")
print(f" The max value of in-state tution of the above four institutions: {tuition_max}")
print(f" The average value of in-state tution of the above four institutions: {tuition_avg}")
```
