## Access to electricity by year in different countries

```python
# <1> 
# Import pandas library
# convert cvs files into one dataframe
# rename columns and make a smaller dataframe of need columns

import pandas as pd 

df = pd.read_csv('/content/gdp_wdi_data_09.csv')
df_2 = pd.read_csv('/content/gdp_wdi_data_18.csv')

df_list = [df,df_2]

df_merged = pd.concat(df_list)
df_merged = df_merged[['Year','SP.POP.TOTL','NY.GDP.PCAP.CD','EG.ELC.ACCS.ZS','Country Code','Country Name']].copy()
df_merged = df_merged.rename(columns={"SP.POP.TOTL": "Total Population", "NY.GDP.PCAP.CD":
                                       "GDP", "EG.ELC.ACCS.ZS": "Access to electricity(% of popuation)", "Country Name": "Country" })
```

```python
# <2> 
# Import plotly express library
# plot and display line chart

import plotly.express as px

fig = px.line(df_merged, x ="Year", y="Access to electricity(% of popuation)",  color="Country", 
  
title="Access to electricity (% of population) by year in different countries")

fig.show()
```



Here is a line chart; different color represents each country.

In 2000, the United States United Kingdom, and surprisingly Mexico had the most access to electricity and continued to have access throughout the years, where is the United States had been leading. The bottom 2 countries in 2000 2 in order, India and Nigeria. Although India and Nigeria have not been stagnant, India has the most growth and been exponentially growing its accessibility over time. Nigeria remains under 60 %, whereas India climbed from 60% to nearly 100% access to electricity.

## GDP(current US$) vs. Accessibility to electricity (% of population) in different countries

```python
# chart 2
# Bubble chart

import plotly.express as px

fig = px.scatter(df_merged, size= 'Total Population', x = "GDP", y="Access to electricity(% of popuation)",  color="Country", hover_name="Country", animation_frame="Year",
  
title="GDP(current US$) vs. Accessibility to electricity (% of population) in different countries")

fig.show()
```

## Mobile Cellular subscriptions in different countries

## CO2 emissions  in Different countries 

```python
##chart 5
# Pie chart
df_5 = pd.read_csv('/content/co2_wdi_data_09.csv')
df_6 = pd.read_csv('/content/co2_wdi_data_18.csv')

df_list = [df_5,df_6]
df_merged = pd.concat(df_list)

df_merged = df_merged[['Year','SP.POP.TOTL','EN.ATM.CO2E.KT','EG.ELC.ACCS.ZS','Country Code','Country Name']].copy()
df_merged_3 = df_merged.rename(columns={"SP.POP.TOTL": "Total Population", "EN.ATM.CO2E.KT":
                                       "CO2 emissions (kt)", "EG.ELC.ACCS.ZS": 
                                     "Access to electricity(% of popuation)", "Country Name": "Country" })
df_merged_3


fig = px.pie(df_merged_3.query("Year == 2000") , values= "CO2 emissions (kt)", names= "Country",  
             color="Country", 
  
title="CO2 emissions in different countries in 2000 ")

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

Above is a pie chart; each pie chart slice represents the country's total percentage of COE emissions.
