## 2.3 Python Language Basics


### Languages Semantics

In Python, use whitespace such as tabs and spaces to structure your code. A colon is followed after a indented code of block, which all 
of the block of code is indented. Whitespaces allows your code to be more readable. Semicolons can be used to seperate multiple code statements on a single line.


### Everything is an object
Everything in Python is a *Python object* with an assocuited *type* and internal data.

### Comments
Texts followed by a hash tag(#) is a comment, and is ignored by Python interpreter. You can also commment out blocks of codes 
using the hash tag symbol.

### Function and object method calls
You can call functiosn using () and passing arguments, and you can assign the return value to a variable. Methods are attached functions to
an objects. You can call a method by using object name dot(.) method and parentheses. 

Example:
```obj.some_method(x,y,z)```
